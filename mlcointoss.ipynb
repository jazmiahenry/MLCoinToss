```python
import numpy as np
from scipy.stats import bernoulli
from scipy.stats import beta
from matplotlib import pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
from IPython.display import IFrame
```

```python
if __name__ == "__main__":
    number_of_trails = [0, 2, 10, 20, 100, 500]
    data = bernoulli.rvs(0.5, size = number_of_trails[-1])
    x = np.linspace(0, 1, 100)
    
    for i, N in enumerate(number_of_trails):
        heads = data[:N].sum()
        
        ax = plt.subplot(len(number_of_trails)/ 2, 2, 1+1)
        ax.set_title("%s trials, %s heads" % (N, heads))
        
        plt.xlabel("Probability of Heads")
        plt.ylabel("Density")
        if i == 0:
            plt.ylim([0.0, 2.0])
        plt.setp(ax.get_yticklabels(), visible=False)
        
        y = beta.pdf(x, 1 + heads, 1 + N - heads)
        plt.plot(x, y, label = "observe %d tosses, \n %d heads" % (N, heads))
        plt.fill_between(x, 0, y, color="red", alpha=0.5)
    fig = plt.gcf()
    fig.set_size_inches(22.5, 15.5) 
    plt.tight_layout()  
    plt.show()
```

```python
if __name__ == "__main__":
    sns.set_palette("deep", desat = .6)
    sns.set_context(rc={"figure.figsize": (20, 12)})
    x = np.linspace(0, 1, 100)
    params = [
        (0.5, 0.5),
        (1, 1),
        (4, 3),
        (2, 5),
        (6, 6)
        ]
    for p in params:
        y = beta.pdf(x, p[0], p[1])
        plt.plot(x, y, label="$\\alpha=%s$, $\\beta=%s$" % p)
    plt.xlabel("$\\theta$, Fairness")
    plt.ylabel("Density")
    plt.legend(title="Parameters")
    fig = plt.gcf()
    fig.set_size_inches(15.5, 9.5) 
    plt.show()
```
